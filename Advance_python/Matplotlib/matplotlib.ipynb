### Why Visualize?

Our brains are wired to process visual information far more effectively than tables of numbers. Good plots can instantly reveal:

- **Trends:** How does a value change over time?
- **Comparisons:** How do different groups compare against each other?
- **Distributions:** How is our data spread out? Are there outliers?
- **Relationships:** Do two variables seem to be related?

Data visualization is the graphical representation of information and data. By using visual elements like charts, graphs, and maps, visualization tools provide an accessible way to see and understand trends, outliers, and patterns in data.


### Python Visualization Libraries

- **Matplotlib:**  
  The foundational plotting library in Python. It's powerful and highly customizable but can be complex for simple plots.

- **Seaborn:**  
  Built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. It often requires less code for beautiful plots.

### Basic Plots with Matplotlib
#### 1. Bar Chart
Excellent for comparing quantities across different categories.

```python
plt.figure(figsize=(6,4))
plt.plot(df['ProductName'],df['Price'],color="Green")
plt.xlabel("Product Name")
plt.title("Comparison between Price and Product Name")
plt.suptitle("Title:")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(8, 4))

plt.bar(df['ProductName'], df['UnitsSold'], color='skyblue')
plt.ylabel("UnitsSold")
plt.xlabel('ProductName')
plt.show()
```

```python
import pandas as pd

df=pd.read_csv('order_cleaned_data.csv')
df
```

```python
plt.figure(figsize=(8, 4))

plt.bar(df['Purchase Date'], df['Total Sales'], color='red')
# plt.ylabel("UnitsSold")
# plt.xlabel('ProductName')
plt.show()
```

```python
plt.figure(figsize=(44,10))
bars = plt.bar(df['Purchase Date'][:100], df['Total Sales'][:100], color='red')
plt.xticks(rotation=90, ha='right') # Rotate x-axis labels for readability
plt.tight_layout()
for bar in bars:
   
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{int(height)}', ha='center', va='bottom')# Adjust plot to prevent labels from overlapping
# plt.xlabel("Product Name")
# plt.title("Comparison between Price and Product Name")
# plt.suptitle("Title:")
# plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(50,10))
bars = plt.bar(df['Product Name'][:100], df['Quantity'][:100], color='b')

plt.title('Comparision')
plt.ylabel('Quantity')
plt.xlabel('Product Name')

plt.xticks(rotation=90, ha='right') # Rotate x-axis labels for readability
plt.tight_layout() # Adjust plot to prevent labels from overlapping

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{int(height)}', ha='center', va='bottom')
    
plt.tight_layout()
plt.show()
```
