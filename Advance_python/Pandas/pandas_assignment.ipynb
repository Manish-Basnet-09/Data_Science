## 3. Application Functions (`.apply`)
**Concepts:** `apply`, `lambda`.

**Task:** Create a column `Order_Size`.
*   If Qty == 1 -> 'Single'
*   If Qty > 1 -> 'Bulk'

```python
def classify_age(age):
    if age < 30:
        return 'Young Adult'
    elif age <= 50:
        return 'Adult'
    else:
        return 'Senior'

# Apply to 'Customer Age' column
df['Age Group'] = df['Customer Age'].apply(classify_age)

# Verify
print(df['Age Group'].value_counts())
display(df[['Customer Age', 'Age Group']].head())
```

```python
y= lambda x:x+(0.13*x)
df['Price with VAT']=df['Price'].apply(y)

display(df[['Product Name', 'Price','Price with VAT']].head())
```

## 3. Merging & Concatenation
### A. Creating Secondary Data (Customer Membership)
Since the main data only has `Customer ID`, we need a separate table to simulate a database join.


```python
unique_customers = df['Customer ID'].unique()
unique_customers
```

```python
# Get unique customers from main df
import numpy as np
import random
unique_customers = df['Customer ID'].unique()

# Assign them random membership levels
membership_data = {
    'Customer ID': unique_customers,
    'Membership Level': np.random.choice(['Bronze', 'Silver', 'Gold'], size=len(unique_customers)),
    'Discount_Eligible': np.random.choice([True, False], size=len(unique_customers))
}

members_df = pd.DataFrame(membership_data)
```

```python
members_df
```

```python
full_data = pd.merge(
    left=df,
    right=members_df,
    on='Customer ID',
    how='left'
)

# Display result
cols_to_show = ['Order ID', 'Total Sales', 'Customer ID', 'Membership Level']
display(full_data[cols_to_show].sample(5))
```

```python
    display(full_data)
```

```python
df (Left Table)          members_df (Right Table)
┌──────────────┐         ┌──────────────┐
│ Customer ID  │         │ Customer ID  │
│──────┬───────│         │──────┬───────│
│ 101  │ Alice │         │ 101  │ Gold  │
│ 102  │ Bob   │         │ 103  │ Silver│
│ 103  │Charlie│         │ 104  │ Bronze│
└──────────────┘         └──────────────┘
```

```python
Result (INNER JOIN):
┌──────────────┬──────────┐
│ Customer ID  │ Membership│
│──────┬───────┼──────────│
│ 101  │ Alice │   Gold   │ ← Only 101 appears in both
│ 103  │Charlie│  Silver  │ ← Only 103 appears in both
└──────────────┴──────────┘
```

```python
Result (LEFT JOIN):
┌──────────────┬──────────┐
│ Customer ID  │ Membership│
│──────┬───────┼──────────│
│ 101  │ Alice │   Gold   │
│ 102  │ Bob   │   NaN    │ ← Bob kept (from left), no match
│ 103  │Charlie│  Silver  │
└──────────────┴──────────┘
```

```python
Result (FULL OUTER JOIN):
┌──────────────┬──────────┐
│ Customer ID  │ Membership│
│──────┬───────┼──────────│
│ 101  │ Alice │   Gold   │
│ 102  │ Bob   │   NaN    │ ← From left only
│ 103  │Charlie│  Silver  │
│ 104  │   NaN │  Bronze  │ ← From right only
└──────────────┴──────────┘
```

```python
Result (CROSS JOIN):
┌──────────────┬──────────┐
│ Customer ID  │ Membership│
│──────┬───────┼──────────│
│ 101  │ Alice │   Gold   │
│ 101  │ Alice │  Silver  │
│ 101  │ Alice │  Bronze  │
│ 102  │ Bob   │   Gold   │
│ 102  │ Bob   │  Silver  │
│ 102  │ Bob   │  Bronze  │
│ 103  │Charlie│   Gold   │
│ 103  │Charlie│  Silver  │
│ 103  │Charlie│  Bronze  │
└──────────────┴──────────┘
```

```python
full_data = pd.merge(
    left=df,
    right=members_df,
    on='Customer ID',
    how='right'
)

# Display result
display(full_data)
```
