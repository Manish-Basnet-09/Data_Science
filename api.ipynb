```python
import requests

a=input("Enter a number:")

api_url = f"https://jsonplaceholder.typicode.com/todos/{a}"

# Make the GET request
response = requests.get(api_url)
print(response)
# The response object contains all the information
print(f"Status Code: {response.status_code}") # 200 means OK

# The actual data is in the response body
# .text gives the raw text, .json() parses it directly into a Python dict
if response.status_code == 200:
    data = response.json()
    print("Data from API:")
    print(data)
    print(f"Title: {data['title']}")
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")
```

```python
import requests

base_currency=input("Enter base currency:")
target_currency= input("Enter target currency: ")
api_url = f"https://open.er-api.com/v6/latest/{base_currency}"
response = requests.get(api_url)

data = response.json()
#print(data)
rate=data['rates']
#print(rate)
def convert_currency(base_currency,target_currency, rate):
    value_of_base = rate[base_currency]
    print(value_of_base)
    value_of_target = rate[target_currency]
    print(value_of_target)

    final_value = value_of_base * value_of_target
    return f"Present value price of {value_of_base} with {value_of_target} : final value"
    
final_value = convert_currency(base_currency,target_currency, rate)
print(final_value)
```

```python
import requests

base_currency=input("Enter base currency:").upper()
target_currency= input("Enter target currency: ").upper()
amount = int(input("Enter the amount you want to convert"))
api_url = f"https://open.er-api.com/v6/latest/{base_currency}"
response = requests.get(api_url)

data = response.json()
#print(data)
rate=data['rates']
#print(rate)
def convert_currency(amount,base_currency,target_currency, rate):
    value_of_base = rate[base_currency]
    #print(value_of_base)
    value_of_target = rate[target_currency]
    #print(value_of_target)

    final_value = amount * value_of_target
    return f"Present value price of {target_currency} with {amount} {base_currency} : {target_currency} {final_value}"
    
final_value = convert_currency(amount,base_currency,target_currency, rate)
print(final_value)
```
