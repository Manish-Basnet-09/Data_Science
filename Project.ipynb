```python
"""Project 1 : Rock, Paper, Scissors 
Goal:  Build a game to play Rock-Paper-Scissors against the computer.
Scenario: A classic game to understand conditional logic (if-elif-else).
Key Concepts:: random.choice(), Boolean logic, Game loops.
Input: User types "Rock", "Paper", or "Scissors".
Output: "Computer chose Rock. You Win!"
"""

import random

print(f"Welcome to  Rock, Paper, Scissor game !!!")

while True:
    user_input=input("Enter your choice Rock, Paper or Scissor (or 'quit' to exit): )")

    if user_input == "quit":
        print(f"Game over! Thank you for playing")
        break
    
    if user_input not in ["rock", "paper", "scissor"]:
        print(f"Invalid choice ! Choose either rock , paper or scissor")
        continue
    
    computer_input = random.choice(["rock", "paper", "scissor"])

    print(f"Computer chooses : {computer_input.capitalize()}")

    #game logic
    if user_input == computer_input:
        print("It's a draw!")

    elif(user_input == "rock" and computer_input == "scissor" or
         user_input == "paper" and computer_input == "rock"or
         user_input == "scissor" and computer_input == "paper"):
        print(f"You Win!!!")

    else:
        print(f"You Loose!!!")


print("-"*80)
```

```python
"""Project 2 : The Quiz Master App 
Goal: A multiple-choice quiz that tracks the user's score.
Scenario: An institute wants to conduct a quick Python entrance test.
Key Concepts: Dictionaries (Key: Question, Value: Answer), JSON handling.
Input:  User selects option A, B, C, or D.
Output: "Correct! Score: 5/10".
"""
```

```python
import json
print("**** Welcome to the Quiz Master App ****")

# JSON string 
quiz_json = '''
{
    "What is the output of 10 % 3?": ["3", "1", "0", "10", "B"],
    "Which keyword is used to define a function in Python?": ["function", "define", "def", "func", "C"],
    "Which data type is immutable?": ["List", "Dictionary", "Set", "Tuple", "D"],
    "Which symbol is used for comments in Python?": ["//", "#", "/*", "--", "B"],
    "What does input() return by default?": ["int", "float", "string", "boolean", "C"],
    "Which operator checks equality?": ["=", "!=", "==", "<>", "C"],
    "Which loop runs a fixed number of times?": ["while", "do-while", "for", "loop", "C"],
    "What is the file extension for Python?": [".pt", ".python", ".p", ".py", "D"],
    "What is the output of type(5)?": ["float", "int", "str", "bool", "B"],
    "Which keyword is used for condition checking?": ["if", "for", "def", "class", "A"]
}
'''

# Convert JSON string to Python dictionary
quiz = json.loads(quiz_json)

score = 0
total_questions = len(quiz)

for question, options in quiz.items():
    print("\n" + question)
    print("A.", options[0])
    print("B.", options[1])
    print("C.", options[2])
    print("D.", options[3])

    user_answer = input("Enter your answer (A/B/C/D): ").upper()

    if user_answer == options[4]:
        score += 1
        print(" Correct!")
    else:
        print(f" Wrong! Correct answer is {options[4]}")

print("\n Quiz Completed")
print(f"Final Score: {score}/{total_questions}")
```

```python
"""Project 3: Secure Password Generator 
Goal: Generate strong, random passwords with specific requirements.
Scenario: An IT admin needs a tool to generate unique passwords for new employees.
Key Concepts: string module, random, List comprehension.
Input:  Length of password, Include symbols (Y/N).
Output: A random string: Xy#9mP$2."""
```

```python
import random
import string

print(" Secure Password Generator ")

# User inputs
length = int(input("Enter password length: "))
include_symbols = input("Include symbols? (Y/N): ").upper()

# Character sets
letters = string.ascii_letters
digits = string.digits
symbols = string.punctuation

# Base pool
char_pool = letters + digits

# Add symbols if selected
if include_symbols == "Y":
    char_pool += symbols

# LIST COMPREHENSION USED HERE ðŸ‘‡
password_list = [random.choice(char_pool) for _ in range(length)]

# Convert list to string
password = "".join(password_list)

print("\nGenerated Password:", password)
```
