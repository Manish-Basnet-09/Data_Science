```python
Objective: Learn NumPy fundamentals
Instructions:

Create a NumPy array of 10 random numbers.

Find:

Mean

Median

Maximum

Minimum

Replace all values greater than mean with 0.

Expected Learning:

NumPy arrays

Mathematical operations

Conditional filtering

```

```python
import numpy as np
import random

# Create NumPy array of 10 random numbers
arr = np.random.randint(1, 100, size=10)

print("Array:", arr)

# Mean
mean_val = np.mean(arr)
print("Mean:", mean_val)

# Median
median_val = np.median(arr)
print("Median:", median_val)

# Maximum
print("Maximum value:", np.max(arr))

#  Minimum
print("Minimum value:", np.min(arr))

# Replace values greater than mean with 0
arr[arr > mean_val] = 0

print("Modified Array:", arr)
```
