```python
#2.Converting the csv data into dataframe
import pandas as pd

df = pd.read_csv('employee_data.csv')
df
```

```python
df = df.set_index('id')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# 3. Remove rows where name is missing
df.dropna(subset=["name"],inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# 4.Fill missing age values with the mean age

mean_age = df['age'].mean()

print(f"The mean age is: {mean_age}")

df.fillna({"age":mean_age},inplace = True)

df
```

```python
df.isnull().sum()
```

```python
# 5. Fill missing salary values with the median salary
median_salary = df['salary'].median()

print(f"The median salary is: {median_salary}")

df.fillna({"salary":median_salary},inplace = True)

df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# 6.Replace missing department values with "Unknown"

df.fillna({"department":"Unknown"},inplace = True)

df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['join_date'].dtype
```

```python
# 7.Convert join_date to a proper datetime format
df["join_date"] = pd.to_datetime(df["join_date"], errors="coerce")
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# 8.Drop rows where join_date is missing

df = df.dropna(subset=["join_date"])
df
```

```python
df.isnull().sum()
```

```python
#9. Forward fill remaining missing values (if any)

df = df.fillna(method="ffill")
df
```

```python
# 10. Ensure age and salary are integers

df["age"] = df["age"].astype(int)
df["salary"] = df["salary"].astype(int)
```

```python
df.info()
```

```python
# 11. Add bonus column (10% of salary)
df["bonus"] = (df["salary"] * 0.10)
df
```

```python
# converting bonus to int
df["bonus"] = df["bonus"].astype(int)
```

```python
df
```

```python
df.info()
```

```python
# Add performance score (1 to 5)
import numpy as np
import random
np.random.seed(0)
df["performance_score"] = np.random.randint(1, 6, size=len(df))
df
```

```python
df=df.reset_index()
df
```

```python
#Renmaing all the headings so that every headings is started with capital letter
df.rename(columns={"id": "Id"}, inplace=True)
df.rename(columns={"name": "Name"}, inplace=True)
df.rename(columns={"age": "Age"}, inplace=True)
df.rename(columns={"salary": "Salary"}, inplace=True)
df.rename(columns={"department": "Department"}, inplace=True)
df.rename(columns={"join_date": "Join Date"}, inplace=True)
df.rename(columns={"bonus": "Bonus"}, inplace=True)
df.rename(columns={"performance_score": "Perfromance Score"}, inplace=True)
df
```
