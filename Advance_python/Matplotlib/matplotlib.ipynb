### Why Visualize?

Our brains are wired to process visual information far more effectively than tables of numbers. Good plots can instantly reveal:

- **Trends:** How does a value change over time?
- **Comparisons:** How do different groups compare against each other?
- **Distributions:** How is our data spread out? Are there outliers?
- **Relationships:** Do two variables seem to be related?

Data visualization is the graphical representation of information and data. By using visual elements like charts, graphs, and maps, visualization tools provide an accessible way to see and understand trends, outliers, and patterns in data.


### Python Visualization Libraries

- **Matplotlib:**  
  The foundational plotting library in Python. It's powerful and highly customizable but can be complex for simple plots.

- **Seaborn:**  
  Built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. It often requires less code for beautiful plots.

##  Quick Decision Guide

| Plot Type   | Best For                          | Example Use Case                          |
|------------|----------------------------------|-------------------------------------------|
| Line Plot  | Trends over time                  | Stock prices, website traffic             |
| Bar Plot   | Comparing categories              | Sales by product, survey results          |
| Scatter Plot | Relationships between variables | Height vs. Weight, Age vs. Income         |
| Histogram  | Distribution of single variable   | Age distribution, exam scores             |
| Box Plot   | Statistical summary               | Comparing salaries across departments     |
| Heatmap    | Matrix / correlation data         | Correlation matrix, confusion matrix      |
| Pie Chart  | Proportions of a whole             | Market share, budget allocation           |

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data - Monthly sales over a year
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2023 = [120, 135, 150, 165, 180, 200,
              190, 210, 220, 230, 250, 300]
sales_2024 = [130, 145, 160, 170, 185, 210,
              200, 220, 235, 245, 265, 320]
sales_2025 = [267,243,362,453,453,345,333,455,367,567,231,132]
plt.figure(figsize=(12, 6))
plt.plot(months, sales_2023, marker='o', linewidth=2, label='2023 Sales', color='blue')
plt.plot(months, sales_2024, marker='s', linewidth=2, label='2024 Sales', color='green', linestyle='-.')
plt.plot(months, sales_2025, marker='D', linewidth=2, label='2025 Sales', color='red', linestyle='--')

plt.suptitle('Monthly Sales Comparison (2023 vs 2024 vs 2025)', fontsize=14, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales (in thousands $)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

```python
# Multiple subplots for different metrics
_, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Simple line plot
axes[0, 0].plot(months, sales_2023)
axes[0, 0].set_title('Simple Line Plot')

# 2. Area plot (filled under line)
axes[0, 1].fill_between(months, sales_2023, alpha=0.3)
axes[0, 1].plot(months, sales_2023, linewidth=2)
axes[0, 1].set_title('Area Plot')
# 3. Step plot
axes[1, 0].step(months, sales_2023, where='pre')
axes[1, 0].set_title('Step Plot')

# 4. Stacked area plot
axes[1, 1].stackplot(months, sales_2023, sales_2024,sales_2025,
                      labels=['2023', '2024','2025'], alpha=0.7)
axes[1, 1].set_title('Stacked Area Plot')
axes[1, 1].legend()

plt.tight_layout()
plt.show()
```

## BAR PLOT

```python
# Sample data - Product sales
products = ['Laptop', 'Phone', 'Tablet', 'Monitor', 'Keyboard']
sales_q1 = [150, 300, 200, 120, 80]
sales_q2 = [180, 320, 210, 140, 90]

x = np.arange(len(products))
width = 0.35


plt.figure(figsize=(10, 6))
bars1 = plt.bar(x + width/2, sales_q1, width,
                label='Q1 Sales', color='red', edgecolor='black')
bars2 = plt.bar(x - width/2, sales_q2, width,
                label='Q2 Sales', color='cyan', edgecolor='black')


plt.title('Product Sales by Quarter', fontsize=14, fontweight='bold')
plt.xlabel('Products', fontsize=12)
plt.ylabel('Units Sold', fontsize=12)
plt.xticks(x, products)
plt.legend()
plt.gca().set_facecolor('#f0f4f8')

plt.yticks(np.arange(0, 361, 20))


# Add value labels on bars
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 5,
                 f'{int(height)}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.grid(True, alpha=0.1)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values1 = [25, 40, 30, 35, 20]
values2 = [30, 35, 1, 40, 30]

# 1. Horizontal bar plot
axes[0, 0].barh(categories, values1, color='cyan')
axes[0, 0].set_title('Horizontal Bar Plot')


# 2. Stacked bar plot
axes[0, 1].bar(categories, values1, label='Group 1', color='cyan')
axes[0, 1].bar(categories, values2, bottom=values1,
               label='Group 2', color='red')
axes[0, 1].set_title('Stacked Bar Plot')
axes[0, 1].legend()


# 3. Grouped bar plot
x = np.arange(len(categories))
width = 0.35
axes[1, 0].bar(x - width/2, values1, width, label='Group 1')
axes[1, 0].bar(x + width/2, values2, width, label='Group 2')
axes[1, 0].set_xticks(x)
axes[1, 0].set_xticklabels(categories)
axes[1, 0].set_title('Grouped Bar Plot')
axes[1, 0].legend()


# 4. Percentage bar plot
total = [v1 + v2 for v1, v2 in zip(values1, values2)]
pct1 = [v1/t*100 for v1, t in zip(values1, total)]
pct2 = [v2/t*100 for v2, t in zip(values2, total)]
axes[1, 1].bar(categories, pct1, label='Group 1 %')
axes[1, 1].bar(categories, pct2, bottom=pct1, label='Group 2 %')
axes[1, 1].set_title('Percentage Bar Plot')
axes[1, 1].set_ylabel('Percentage (%)')
axes[1, 1].legend()

plt.tight_layout()
plt.show()
```

```python


# Sample data - Study Hours vs Exam Scores
np.random.seed(42)
study_hours = np.random.normal(15, 5, 100)

print(study_hours)
exam_scores = 50 + 2*study_hours + np.random.randint(0, 10, 100)
# Add some noise and relationship

plt.figure(figsize=(10, 6))
scatter = plt.scatter(study_hours, exam_scores,
                     c=exam_scores, cmap='viridis',
                     s=100, alpha=0.7, edgecolors='black')

plt.title('Study Hours vs Exam Scores', fontsize=14, fontweight='bold')
plt.xlabel('Study Hours per Week', fontsize=12)
plt.ylabel('Exam Score (out of 100)', fontsize=12)
plt.colorbar(scatter, label='Exam Score')

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
