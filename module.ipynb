```python
import datetime as dt

now = dt.datetime.now()
print(f"Current date and time: {now}")
print(f"Current year: {now.year}")
now.strftime("%Y-%m-%d, %H:%M:%S %p, %A")
print(type(now.strftime("%Y-%m-%d, %H:%M:%S %p, %A")))
```
