```python
"""Task 1: The
"Cyber-ID" Generator

Scenario: You are building a registration system
for a futuristic gaming convention.

Instructions:

1. Ask the user for
their username, birth_year (as integer),
and is_member (True/False).

2. Calculate their
current age (assume current year is 2025).

3. Create a
variable entry_fee. If they are a member, fee is 10.5,
otherwise 20.0.

4. Output: Print
a formatted ID card:

`"User: [Name]"""

#Cyber ID generator program
#taking the user input
username=input("Enter your name: ")
birth_year=int(input("Enter your birth year: "))
is_member_input=input("Are you the member of the cyber?(True/False): ").strip().lower()

#converting string values into boolean
is_member = is_member_input == "true"

#for age calculations
current_year = 2025
age = current_year - birth_year

#entry fee determination 
if is_member:
    entry_fee=10.5
else:
    entry_fee=20.0

#print the data for the id card
print("\n-------CYBER ID CARD-------")
print(f"{'Name':<15}: {username}")
print(f"{'Age':<15}: {age}")
print(f"{'Member':<15}: {is_member}")
print(f"{'Entry fee':<15}: Rs.{entry_fee:.2f}")
```

```python
"""Assignment

Activity 1: Number Guessing Game
Goal: The computer picks a secret number, and the user has to guess it. The program gives hints like "too high" or "too low".
Concepts you can use:
import random (to get a random number)
input() (to get the user's guess)
int() (to convert the user's input string into a number)
while loop (to keep guessing until correct)
if/elif/else (to check the guess and give hints)"""

import random

# Computer picks a secret number between 1 and 100
secret_number = random.randint(1, 100)

# Initialize guess
guess = 0

print("Guess the number between 1 and 100")

# Loop until the guess is correct
while guess != secret_number:
    guess = int(input("Enter your guess: "))

    if guess > secret_number:
        print("Too high!")
    elif guess < secret_number:
        print("Too low!")
    else:
        print("Congratulations! You guessed the correct number.")
```

```python
"""Scenario: HR needs a reusable tool to calculate monthly salary based on hours worked and tax.

Instructions:

1. Define a function calculate_salary(hours, hourly_rate).

2. Logic: Calculate gross = hours * hourly_rate.

3. Apply Tax: If gross > 5000, deduct 10%. Otherwise, deduct 5%.

4. Return the final net salary.

5. Call the function twice with different values (e.g., John worked 40hrs @ 100, Jane worked 60hrs @ 200) and print the results."""

#The Salary Calculator

#function for calculating the net salary
def calculate_salary(hours, hourly_rate):
    gross = hours * hourly_rate

    if gross > 5000:
        tax = ((10/100) * gross)

    else:
        tax = ((5/100) * gross)

    net_salary = gross - tax

    return net_salary

#main function to print salary of john and jane
#function call
John_salary = calculate_salary(40, 100)
Jane_salary = calculate_salary(60, 200)

#display salary
print(f"\nNet salary of John is {John_salary}")
print(f"\nNet salary of Jane is {Jane_salary}")
```

```python
"""Scenario: A fitness app needs to calculate the total steps taken in a week and find the average.

Instructions:

Create an empty list called steps.

Ask the user to input the number of steps taken for 7 days and store each value in the list. Hint: use range function for 7 times user input

Create a variable total_steps = 0.

Loop:

Iterate through the list.

Add each day’s steps to total_steps.

Print "Day [X]: [Steps] steps".

After the loop:

Calculate the average steps taken in the week (total_steps / 7).

Print:

Weekly Total: [Total]"""

#A fitness app

#an empty list steps is created
steps = []

#taking input from the user using range function
for day in range (1,8):
    daily_steps = int(input(f"Enter steps for Day {day} "))
    steps.append(daily_steps)

#intialize total steps 
total_steps = 0

#loop through lists
for i in range (len(steps)):
    total_steps = total_steps + steps[i]
    print(f"Day {i+1} : {steps[i]} steps")

#calculate average steps
average_steps = total_steps / 7

#print the results
print(f"\nWeekly Total : {total_steps}")
print(f"Average Steps : {average_steps}")

```

```python
"""Scenario: A security door requires a PIN code. It allows 3 attempts before locking out.

Instructions:

Set a secret PIN: secret = 1234.
Set attempts = 0.

Start a while loop.

Ask the user to "Enter PIN".
Increase attempts by 1.

Conditions:

If the input matches secret:
Print "Access Granted" and break.

If attempts reaches 3:
Print "System Locked" and break.

Else:
Print "Wrong PIN, try again."

"""

#initialize secret and attempts
secret = 1234
attempts = 0

#use while loops
while True:
    pin = int(input("Enter the PIN: "))
    attempts += 1

    if pin == secret:
        print("Access granted!")
        break

    elif attempts == 3:
        print("System Locked!")
        break
        
    else:
        print("Wrong PIN, Try Again")

        
```

```python
"""Scenario: You intercepted a scrambled message from a rival spy agency. You need to decode it.

Instructions:

1. Variable scrambled = "##gnimmargorP nohtyP evoL I##".

2. Clean: Remove the hashtags # using .strip() or .replace().

3. Reverse: Flip the string so it reads correctly (Hint: slicing [::-1]).

4. Analyze: Count how many times the letter "o" appears in the decoded text.

5. Print the final clean message and the count."""

#The Spy Message Decoder

#initial stirng
scrambled = "##gnimmargorP nohtyP evoL I##"

#cleaned string after replaceing '#' with ''
cleaned = scrambled.replace("#", "")

#reversed string 
reversed_string = cleaned[::-1]

#count the letter "O" in the string
count = reversed_string.count("o")

#print the result
print(f"\nThe clean decoded message is: {reversed_string}")
print(f"\nCount of letter 'o' in the message is: {count}")
```

```python
"""Scenario: You have a pile of returned books (some duplicates) and need to update the library catalog.

Instructions:

1. returned_books = [

"1984", "Hobbit", "1984", "Dune", "Hobbit",

"Foundation", "Dune", "Brave New World",

"1984", "Fahrenheit 451", "Hobbit"

]

2. Unique: Convert the list to a Set to remove duplicates.

3. Catalog: By using the for loop in Create a dictionary library = {'1984': 5, 'Hobbit': 2, "Dune":4}.

4. Update: Loop through the unique set. If the book is in library, add +1 to its count. If not, add it with count 1.

5. Print the final library dictionary.
"""
#Library Book Organizer

#initial book list
returned_books = [

"1984", "Hobbit", "1984", "Dune", "Hobbit",

"Foundation", "Dune", "Brave New World",

"1984", "Fahrenheit 451", "Hobbit"

]

#converting book list to set
unique_books = set(returned_books)

#creating Dictionary library using for loop

library={}
for i in range(3):
    key=str(input("\nEnter key "))
    value=int(input("\nEnter value "))

    library[key] = value

#updating the books in library
for book in unique_books:
    if book in library:
        library[book] += 1
    else:
        library[book] = 1

#print final catalog
print(f"\nFinal library catalog : \n{library}")
```

```python
#Simple class concept

"""Design a class Student with attributes name, roll_number, and marks
(list of 5 subject marks). Write a method to calculate the average marks
and grade of the student (grade based on average: A ≥ 90, B ≥ 75, C ≥ 50, F
< 50). Test your class with one student object."""

class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no =roll_no
        self.marks = marks
        print(f"A new Student object of {self.name} was created!")
    def calculate_average(self):   # Method to calculate average marks 
        total = 0
        for mark in self.marks:
            total += mark
        average = total / len(self.marks)
        return average   
    def calculate_grade(self):         # Method to determine grade
        average = self.calculate_average()
        if average >= 90:
            return "A"
        elif average >= 75:
            return "B"
        elif average >= 50:
            return "C"
        else:
            return "F"

s1 = Student(
    name="Manish",
    roll_no=12,
    marks=[85, 90, 78, 88, 92]
)

# Display results
avg = s1.calculate_average()
grade = s1.calculate_grade()

print(f"\nStudent Information")
print(f"Name        : {s1.name}")
print(f"Roll Number : {s1.roll_no}")
print(f"Average     : {avg}")
print(f"Grade       : {grade}")
```
